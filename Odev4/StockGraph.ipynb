# Importing Libraries

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# Choose Stock

```python
symbol = "TSLA"
data = yf.download(symbol, start="2023-01-01", end="2024-10-23")
```

# 20 day SMA and  50 day EMA

```python
# Calculate 20 day SMA
data['SMA_20'] = data['Close'].rolling(window=20).mean()

# Calculate 50 day EMA 
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()
```

# Graph Drawing

```python
# Creating Graph
plt.figure(figsize=(10,6))
plt.plot(data['Close'], label='Closing Price', color='blue')
plt.plot(data['SMA_20'], label='20 Day SMA', color='green')
plt.plot(data['EMA_50'], label='50 Day EMA', color='red')

# Title and Tags
plt.title(f'{symbol} - 20-Day SMA and 50-Day EMA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
```
